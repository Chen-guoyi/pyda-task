### 处理缺失数据

```python
# 对于数值数据 ，pandas 使用浮点数值 NaN (Not a Number) 表示缺失数据
import pandas as pd
from pandas import Series
import numpy as np

string_data = pd.Series(['aardvark', 'artichoke', np.nan, 'avocado'])
string_data
```

```python
import pandas as pd
from pandas import Series
import numpy as np

string_data = pd.Series(['aardvark', 'artichoke', np.nan, 'avocado'])
string_data.isnull()
```

```python
# 将缺失值表示为 NA，它表示不可用 not available。
# 在统计应用中，NA 数据可能是不存在的数据或者虽然存在，但是没有观察到（例如，数据采集中发生了问题）。
# 当进行数据清洗以进行分析时，最好直接对缺失数据进行分析，以判断数据采集的问题或缺失数据可能导致的偏差。
# Python 内置的 None 值在对象数组中也可以作为 NA:
import pandas as pd
from pandas import Series
import numpy as np

string_data = pd.Series(['aardvark', 'artichoke', np.nan, 'avocado'])
string_data[0] = None
string_data.isnull()
```

**缺失数据处理的函数**

| 方法  | 说明  |
| ----- | ----- |
| dropna | 根据各标签的值中是否存在缺失数据对轴标签进行过滤，可通过阈值调节对缺失值的容忍度 |
| fillna | 用指定值或插值方法（如 ffill 或 bfill）填充缺失数据 |
| isnull | 返回一个含有布尔值的对象，这些布尔值表示哪些值是缺失值/NA，该对象的类型与源类型一样 |
| notnull | isnull 的否定式 |

#### 填充缺失数据

```python
# 你可能不想滤除缺失数据（有可能会丢弃跟它有关的其他数据），而是希望通过其他方式填补那些“空洞”。
# 对于大多数情况而言，fillna 方法是最主要的函数。
# 通过一个常数调用 fillna 就会将缺失值替换为那个常数值：

import pandas as pd
from pandas import DataFrame
import numpy as np

df = pd.DataFrame(np.random.randn(7, 3))
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
df.fillna(0)
```

```python
# 若是通过一个字典调用 fillna，就可以实现对不同的列填充不同的值

import pandas as pd
from pandas import DataFrame
import numpy as np

df = pd.DataFrame(np.random.randn(7, 3))
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
df.fillna({1: 0.5, 2: 0})
```

```python
# fillna默认会返回新对象，但也可以对现有对象进行就地修改：

import pandas as pd
from pandas import DataFrame
import numpy as np

df = pd.DataFrame(np.random.randn(7, 3))
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
_ = df.fillna(0, inplace=True)
df
```

```python
# 对reindexing有效的那些插值方法也可用于fillna：
import pandas as pd
from pandas import DataFrame
import numpy as np

df = pd.DataFrame(np.random.randn(6, 3))
df.iloc[2:, 1] = NA
df.iloc[4:, 2] = NA
df
```

```python
# 对reindexing有效的那些插值方法也可用于fillna：
import pandas as pd
from pandas import DataFrame
import numpy as np

df = pd.DataFrame(np.random.randn(6, 3))
df.iloc[2:, 1] = NA
df.iloc[4:, 2] = NA
df.fillna(method='ffill')
```

```python
# 对reindexing有效的那些插值方法也可用于fillna：
import pandas as pd
from pandas import DataFrame
import numpy as np

df = pd.DataFrame(np.random.randn(6, 3))
df.iloc[2:, 1] = NA
df.iloc[4:, 2] = NA
df.fillna(method='ffill', limit=2)
```

```python
# 可以利用fillna实现许多别的功能。比如说，你可以传入Series的平均值或中位数：
import pandas as pd
from pandas import Series

data = pd.Series([1., NA, 3.5, NA, 7])
data.fillna(data.mean())
```

>**fillna 参数参考表**

| 参数 | 说明 |
| ---- | ---- |
| value | 用于填充缺失值的标量值或字典对象 |
| method | 插值方式。如果函数调用时未指定其他参数的话，默认为 "ffill" |
| axis | 待填充的轴，默认 axis=0 |
| inplace | 修改调用者对象而不产生副本 |
| limit | （对于前向和后向填充）可以连续填充的最大数量 |

#### 数据转换

##### 移除重复数据

```python
import pandas as pd
from pandas import DataFrame

data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                    'k2': [1, 1, 2, 3, 3, 4, 4]})
data
```

```python
# DataFrame的duplicated方法返回一个布尔型Series，表示各行是否是重复行（前面出现过的行）：
import pandas as pd
from pandas import DataFrame

data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                    'k2': [1, 1, 2, 3, 3, 4, 4]})
data.duplicated()
```

```python
# 还有一个与此相关的drop_duplicates方法，它会返回一个DataFrame：
import pandas as pd
from pandas import DataFrame

data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                    'k2': [1, 1, 2, 3, 3, 4, 4]})
data.drop_duplicates()
```

```python
# 可以指定部分列进行重复项判断。假设我们还有一列值，且只希望根据 k1 列过滤重复项：
import pandas as pd
from pandas import DataFrame

data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                    'k2': [1, 1, 2, 3, 3, 4, 4]})
data['v1'] = range(7)
data.drop_duplicates(['k1'])
```

```python
# duplicated和drop_duplicates默认保留的是第一个出现的值组合。
# 传入 keep='last' 则保留最后一个：
import pandas as pd
from pandas import DataFrame

data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                    'k2': [1, 1, 2, 3, 3, 4, 4]})
data['v1'] = range(7)
data.drop_duplicates(['k1', 'k2'], keep='last') # 索引 5 被删除
```

##### 利用函数或映射进行数据转换

```python
import pandas as pd
from pandas import DataFrame

data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon', 
                              'pastrami', 'honey ham', 'nova lox'],
                    'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
# 添加一列表示该肉类食物来源的动物类型
# 先编写一个不同肉类到动物的映射：
meat_to_animal = {
    'bacon': 'pig',
    'pulled pork': 'pig',
    'pastrami': 'cow',
    'corned beef': 'cow',
    'honey ham': 'pig',
    'nova lox': 'salmon'
}
```

```python
# Series 的 str.lower 方法将各个值转换为小写
import pandas as pd
from pandas import DataFrame

data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon', 
                              'pastrami', 'honey ham', 'nova lox'],
                    'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
lowercased = data['food'].str.lower()
lowercased
```

```python
# Series 的 map 方法可以接受一个函数或含有映射关系的字典对象
import pandas as pd
from pandas import DataFrame

data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon', 
                              'pastrami', 'honey ham', 'nova lox'],
                    'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
lowercased = data['food'].str.lower()

meat_to_animal = {
    'bacon': 'pig',
    'pulled pork': 'pig',
    'pastrami': 'cow',
    'corned beef': 'cow',
    'honey ham': 'pig',
    'nova lox': 'salmon'
}

data['animal'] = lowercased.map(meat_to_animal)
data
```

```python
# 
import pandas as pd
from pandas import DataFrame

data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                              'Pastrami', 'corned beef', 'Bacon', 
                              'pastrami', 'honey ham', 'nova lox'],
                    'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})

meat_to_animal = {
    'bacon': 'pig',
    'pulled pork': 'pig',
    'pastrami': 'cow',
    'corned beef': 'cow',
    'honey ham': 'pig',
    'nova lox': 'salmon'
}

data['food'].map(lambda x: meat_to_animal[x.lower()])
```

#### 替换值

```python
# replace 修改对象的数据子集 更简单、更灵活
import pandas as pd
from pandas import Series

data = pd.Series([1., -999., 2., -999., -1000., 3.])
data
```

```python
# -999 可能是一个表示缺失数据的标记值。要将其替换为 pandas 能够理解的 NA值
# 可以利用 replace 来产生一个新的 Series (除非传入 inplace=True)
import pandas as pd
from pandas import Series
import numpy as np

data = pd.Series([1., -999., 2., -999., -1000., 3.])
data.replace(-999, np.nan)
```

```python
# 如果希望一次性替换多个值，可以传入一个由待替换值组成的列表以及一个替换值
import pandas as pd
from pandas import Series
import numpy as np

data = pd.Series([1., -999., 2., -999., -1000., 3.])
data.replace([-999, -1000], np.nan)
```

```python
# 要让每个值有不同的替换值，可以传递一个替换列表
import pandas as pd
from pandas import Series
import numpy as np

data = pd.Series([1., -999., 2., -999., -1000., 3.])
data.replace([-999, -1000], [np.nan, 0])
```

```python
# 传入的参数也可以是字典
import pandas as pd
from pandas import Series
import numpy as np

data = pd.Series([1., -999., 2., -999., -1000., 3.])
data.replace({-999: np.nan, -1000: 0})
```

#### 重命名索引

```python
import pandas as pd
from pandas import DataFrame
import numpy as np

data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                   index=['Ohio', 'Colorado', 'New York'],
                   columns=['one', 'two', 'three', 'four'])
data
```

```python
# 轴标签也可以通过函数或映射进行转换，而得到一个新的不同标签的对象
# 轴还可以被就地修改，而无需新建一个数据结构

import pandas as pd
from pandas import DataFrame
import numpy as np

data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                   index=['Ohio', 'Colorado', 'New York'],
                   columns=['one', 'two', 'three', 'four'])

# 跟Series一样，轴索引也有一个map方法：
transform = lambda x: x[:4].upper()
data.index.map(transform)
```

```python
# 给 index 赋值，就可以对 DataFrame 进行就地修改
import pandas as pd
from pandas import DataFrame
import numpy as np

data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                   index=['Ohio', 'Colorado', 'New York'],
                   columns=['one', 'two', 'three', 'four'])

transform = lambda x: x[:4].upper()
data.index = data.index.map(transform)
data
```

```python
# 如果想要创建数据集的转换版（而不是修改原始数据），使用 rename
import pandas as pd
from pandas import DataFrame
import numpy as np

data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                   index=['Ohio', 'Colorado', 'New York'],
                   columns=['one', 'two', 'three', 'four'])
data.rename(index=str.title, columns=str.upper)
```

```python
# rename 可以结合字典型对象实现部分轴标签的更新
import pandas as pd
from pandas import DataFrame
import numpy as np

data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                   index=['Ohio', 'Colorado', 'New York'],
                   columns=['one', 'two', 'three', 'four'])
transform = lambda x: x[:4].upper()
data.index = data.index.map(transform)
data.rename(index={'OHIO': 'INDIANA'},
           columns={'three': 'peekaboo'})
```

```python
# rename 可以实现复制 DataFrame 并对其索引和列标签进行赋值
# 若希望就地修改某个数据集，传入 inplace=True 即可
import pandas as pd
from pandas import DataFrame
import numpy as np

data = pd.DataFrame(np.arange(12).reshape((3, 4)),
                   index=['Ohio', 'Colorado', 'New York'],
                   columns=['one', 'two', 'three', 'four'])
transform = lambda x: x[:4].upper()
data.index = data.index.map(transform)
data.rename(index={'OHIO': 'INDIANA'}, inplace=True)
data
```

#### 排列和随机采样

```python
# 利用 numpy.random.permutation 函数可以轻松实现对 Series 或 DataFrame 的列的排列工作
# 通过需要排列的轴的长度调用 permutation，可产生一个表示新顺序的整数数组：
import numpy as np
import pandas as pd
from pandas import DataFrame

df = pd.DataFrame(np.arange(5 * 4).reshape((5, 4)))
sampler = np.random.permutation(5)
sampler
```

```python
import numpy as np
import pandas as pd
from pandas import DataFrame

df = pd.DataFrame(np.arange(5 * 4).reshape((5, 4)))
df
```

```python
# 在基于 iloc 的索引操作或 take 函数中使用该数组
import numpy as np
import pandas as pd
from pandas import DataFrame

df = pd.DataFrame(np.arange(5 * 4).reshape((5, 4)))
sampler = np.random.permutation(5)
df.take(sampler)
```

```python
# 如果不想用替换的方式选取随机子集，可以在 Series 和 DataFrame 上使用 sample 方法：
import numpy as np
import pandas as pd
from pandas import DataFrame

df = pd.DataFrame(np.arange(5 * 4).reshape((5, 4)))
sampler = np.random.permutation(5)
df.sample(n=3)
```

```python
# 要通过替换的方式产生样本（允许重复选择），可以传递 replace=True 到 sample:
import pandas as pd
from pandas import Series

choices = pd.Series([5, 7, -1, 6, 4])
draws = choices.sample(n=10, replace=True)
draws
```
